#Question 1: What is Simple Linear Regression?
   Simple linear regression is a statistical method used to model the relationship between one independent variable and one dependent variable using a straight line. It aims to find the "best fit" line through a scatter plot of data points, minimizing the sum of squared differences between the actual data points and the predicted values on the line. This line can then be used to predict the value of the dependent variable based on the independent variable.
Top Linear Regression Interview Questions

Here's a more detailed breakdown:

One Independent Variable:

Simple linear regression focuses on the relationship between a single predictor variable (also known as the independent variable, input, or feature) and a response variable (also known as the dependent variable, output, or target).
Linear Relationship:
It assumes a linear relationship, meaning the relationship can be represented by a straight line.
Best Fit Line:
The goal is to find the line that best represents the relationship in the data, minimizing the overall error between the predicted values and the actual values.
Equation:
The equation of the line is typically represented as Y = a + bX, where:
Y is the dependent variable.
X is the independent variable.
a is the y-intercept (the point where the line crosses the y-axis).
b is the slope of the line (how much Y changes for each unit change in X).



#Question 2: What are the key assumptions of Simple Linear Regression?
   The key assumptions of simple linear regression are linearity, independence, homoscedasticity, and normality. These assumptions ensure that the model accurately reflects the relationship between the independent and dependent variables and that the results are reliable.
Here's a breakdown of each assumption:
1. Linearity:
A linear relationship exists between the independent variable (X) and the dependent variable (Y). In other words, the relationship can be approximated by a straight line.
2. Independence:
The residuals (the differences between observed and predicted values) are independent of each other. This means that the error for one observation does not influence the error for any other observation.
3. Homoscedasticity:
The variance of the residuals is constant across all levels of the independent variable. In simpler terms, the spread of the data points around the regression line should be roughly the same for all values of X.
4. Normality:
For any given value of X, the dependent variable Y is normally distributed. This means that the distribution of the errors is also normal.
These assumptions are crucial for the validity of the linear regression model. If these assumptions are violated, the model's predictions may be inaccurate, and the statistical tests performed on the model may not be reliable.

#Question 3: What is heteroscedasticity, and why is it important to address in regression models?
   Heteroscedasticity refers to the unequal variance of errors (residuals) across different levels of an independent variable in a regression model. In simpler terms, it means the spread of the data points around the regression line is not consistent; some parts of the data might be more scattered than others. Addressing heteroscedasticity is crucial because it can lead to unreliable statistical inferences and inaccurate predictions.
Why is it important to address?
1. Violates OLS assumptions:
Standard Ordinary Least Squares (OLS) regression assumes homoscedasticity, meaning constant variance of errors. Heteroscedasticity violates this assumption, rendering the standard OLS estimates of standard errors and p-values invalid.
2. Biased standard errors and p-values:
Heteroscedasticity can lead to underestimated standard errors, which inflates the t-statistics and leads to incorrect rejection of the null hypothesis, resulting in false positives.
3. Inefficient parameter estimates:
While the estimates of the regression coefficients might still be unbiased in the presence of heteroscedasticity, they are not the most efficient estimates, meaning they don't have the smallest variance possible.
4. Unreliable predictions:
Due to the biased standard errors and inefficient parameter estimates, the predictions made by the model might not be reliable, especially in areas of the data where heteroscedasticity is pronounced.
In essence, heteroscedasticity can lead to misleading conclusions about the significance of the independent variables and the overall fit of the model, making it essential to detect and address it during regression analysis.

#Question 4: What is Multiple Linear Regression?
   Multiple linear regression is a statistical method that analyzes the relationship between one dependent variable and two or more independent variables. It aims to find the best-fitting linear equation that predicts the dependent variable based on the values of the independent variables. Essentially, it's an extension of simple linear regression, allowing for more than one predictor in the model.
Here's a breakdown:
Dependent Variable:
The variable you are trying to predict (e.g., house price).
Independent Variables:
The variables used to predict the dependent variable (e.g., square footage, number of bedrooms, location).
Linear Relationship:
The assumption that the relationship between the independent and dependent variables can be approximated by a straight line (or a hyperplane in multiple dimensions).
Goal:
To find the coefficients (slopes and intercepts) of the linear equation that best fit the observed data and allow for accurate predictions.
In simpler terms: Imagine you want to predict how well a student will do in a class. Simple linear regression would only use one factor, like the student's previous test scores. Multiple linear regression, on the other hand, could use multiple factors like test scores, study hours, and attendance to make a more accurate prediction.
Key Applications:
Predictive Modeling:
Predicting outcomes based on multiple factors (e.g., predicting a customer's purchase based on demographics and past behavior).
Identifying Important Predictors:
Determining which independent variables have the strongest influence on the dependent variable.
Understanding Relationships:
Gaining insights into how multiple factors interact to affect a specific outcome.

#Question 5: What is polynomial regression, and how does it differ from linear regression?
   Polynomial regression is a type of regression analysis that models the relationship between the independent variable (input) and the dependent variable (output) using an nth-degree polynomial equation. Unlike linear regression, which assumes a linear relationship and fits a straight line, polynomial regression can fit curves and other non-linear relationships in the data.
Key Differences:
Linear Regression:
Uses a linear equation (y = mx + b) to model the relationship, fitting a straight line to the data. It is suitable for data with a clear linear trend.
Polynomial Regression:
Employs a polynomial equation (e.g., y = ax² + bx + c) to model the relationship. This allows it to fit curves and capture more complex patterns in the data.
How Polynomial Regression Works:
1. Extending Linear Regression:
Polynomial regression can be seen as an extension of linear regression, where higher-order terms (e.g., x², x³, etc.) are added to the linear equation to allow for curves.
2. Finding the Best Fit:
The goal is to find the polynomial equation that best fits the data points, minimizing the difference between the predicted and actual values (often using the least squares method).
3. Degree of the Polynomial:
The "degree" of the polynomial (the highest power of the independent variable) determines the complexity of the curve that can be fit. A higher degree allows for more complex curves but also increases the risk of overfitting.
When to Use Polynomial Regression:
When the relationship between variables is not linear and a curved line is needed to fit the data.
When the data shows a clear curvilinear pattern.
Risk of Overfitting:
A key concern with polynomial regression is the risk of overfitting, especially with high-degree polynomials. Overfitting occurs when the model learns the training data too well, including noise, and performs poorly on new, unseen data.
In essence, polynomial regression provides a more flexible approach to modeling relationships than linear regression, especially when dealing with non-linear data, but it requires careful consideration of the degree of the polynomial to avoid overfitting.

#Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy import stats  # optional

# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# === Method 1: NumPy polyfit (least squares) ===
m_np, b_np = np.polyfit(X, Y, deg=1)
print(f"NumPy polyfit → slope = {m_np:.4f}, intercept = {b_np:.4f}")

# === Method 2 (optional): SciPy stats.linregress ===
slope_s, intercept_s, rval, pval, stderr = stats.linregress(X, Y)
print(f"SciPy linregress → slope = {slope_s:.4f}, intercept = {intercept_s:.4f}, R² = {rval**2:.4f}")

# === Method 3: scikit‑learn LinearRegression ===
X_reshaped = X.reshape(-1, 1)
model = LinearRegression()
model.fit(X_reshaped, Y)
m_sk = model.coef_[0]
b_sk = model.intercept_
print(f"sklearn LinearRegression → slope = {m_sk:.4f}, intercept = {b_sk:.4f}")

# === Plotting ===
plt.scatter(X, Y, color='blue', label='Data points')
x_line = np.linspace(X.min(), X.max(), 100)
y_line = m_sk * x_line + b_sk
plt.plot(x_line, y_line, color='red', label=f'Regression line: y = {m_sk:.3f}x + {b_sk:.3f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data:
#● Area = [1200, 1500, 1800, 2000]
#● Rooms = [2, 3, 3, 4]
#● Price = [250000, 300000, 320000, 370000]
#Check for multicollinearity using VIF and report the results.
#(Include your Python code and output in the code box below.)


import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Prepare X, y
X = df[['Area', 'Rooms']]
y = df['Price']

# Add constant for intercept
X_const = sm.add_constant(X)

# Fit multiple linear regression
model = sm.OLS(y, X_const).fit()
print(model.summary())

# Compute VIF
vif_data = pd.DataFrame()
vif_data['feature'] = X_const.columns
vif_data['VIF'] = [variance_inflation_factor(X_const.values, i)
                   for i in range(X_const.shape[1])]
print("\nVariance Inflation Factor (VIF) for predictors:")
print(vif_data)
```

```python
#Question 8: Implement polynomial regression on the following data:
#● X = [1, 2, 3, 4, 5]
#● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
#Fit a 2nd-degree polynomial and plot the resulting curve

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform features to include x and x^2 terms
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
# X_poly now has columns [x, x^2]

# Fit linear regression to polynomial features
model = LinearRegression()
model.fit(X_poly, Y)

# Extract learned coefficients
coef1, coef2 = model.coef_
intercept = model.intercept_
print(f"Model: y = {intercept:.4f} + {coef1:.4f} x + {coef2:.4f} x^2")

# Evaluate fit quality
y_pred = model.predict(X_poly)
r2 = r2_score(Y, y_pred)
print(f"R² score = {r2:.4f}")

# Plot data and fitted curve
plt.scatter(X, Y, color='blue', label='Data points')
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_line_poly = poly.transform(X_line)
Y_line = model.predict(X_line_poly)
plt.plot(X_line, Y_line, color='red', label='Quadratic fit (degree=2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2nd‑Degree Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data:
#● X = [10, 20, 30, 40, 50]
#● Y = [15, 35, 40, 50, 65]
#Assess heteroscedasticity by examining the spread of residuals.

import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns

# Sample data
X = np.array([10, 20, 30, 40, 50])
Y = np.array([15, 35, 40, 50, 65])

# Add constant for intercept
X_const = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X_const).fit()

# Extract fitted values and residuals
fitted = model.fittedvalues
residuals = model.resid

# Standardized residuals
infl = model.get_influence()
resid_std = infl.resid_studentized_internal

# Residuals vs Fitted plot
plt.figure(figsize=(6,4))
sns.residplot(x=fitted, y=residuals, lowess=True,
              scatter_kws={'alpha':0.7},
              line_kws={'color':'red', 'lw':1})
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.show()

# Scale-Location (Spread–Location) plot
plt.figure(figsize=(6,4))
sns.regplot(x=fitted, y=np.sqrt(np.abs(resid_std)), lowess=True,
            scatter_kws={'alpha':0.7},
            line_kws={'color':'red', 'lw':1})
plt.xlabel('Fitted values')
plt.ylabel(r'$\sqrt{|Standardized\ Residuals|}$')
plt.title('Scale-Location Plot')
plt.show()

# Print summary
print(model.summary())
```
